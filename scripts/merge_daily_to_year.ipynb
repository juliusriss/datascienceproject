```python
import os

"""
Read all files in a folder and combine them into a single dataframe.
Get the date from the last 14 characters of the file name.
"""

folder_path = '../default_data/daily_global_charts/daily2017'

all_data = []

for file in os.listdir(folder_path):
    file_path = os.path.join(folder_path, file)
    date_str = file[-14:-4]
    df = pd.read_csv(file_path)
    df['date'] = date_str
    all_data.append(df)

final_df = pd.concat(all_data, ignore_index=True)

final_df['date'] = pd.to_datetime(final_df['date'], format='%Y-%m-%d')
final_df = final_df.sort_values(by='date').reset_index(drop=True)

```

```python
final_df.tail(201)
```

```python

final_df.to_csv('daily2017.csv', index=False)
```
