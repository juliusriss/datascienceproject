# This Notebook merges the csv files for any country into one big csv and adds the date column from the file names as well as the country name if needed.

```python

import pandas as pd
import os
import chardet


#Testing on 1 file

#file_path = '../data/raw_data/daily_per_country_charts/ger-daily'
#date_str = file_path[-14:-4]
#df = pd.read_csv(file_path)
#df['date'] = date_str

#df.tail()
```

```python
folder_path = '../data/raw_data/daily_per_country_charts/in_daily_24'
all_data = []

for file in os.listdir(folder_path):
    if file.endswith('.csv'):
        file_path = os.path.join(folder_path, file)
        date_str = file[-14:-4]

        # Read file using chardet for encoding detection
        with open(file_path, 'rb') as f:
            result = chardet.detect(f.read(10000))  # Read first 10KB
            encoding = result['encoding']

        try:
            df = pd.read_csv(file_path, encoding=encoding)
            df['date'] = date_str
            all_data.append(df)
        except Exception as e:
            print(f"Error reading {file}: {e}")

final_df = pd.concat(all_data, ignore_index=True)
```
